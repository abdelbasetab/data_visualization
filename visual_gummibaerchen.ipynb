```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
np.random.seed(0)
gummybears_2024 = pd.read_csv("gummybears - 2024.csv")
```

```python
gummybears_2024.head()
```

# 2. data cleaning 

```python
missing_values_count = gummybears_2024.isnull().sum()


missing_values_count[0:]
```

```python
# in percent

total_cells = np.prod(gummybears_2024.shape)
total_missing = missing_values_count.sum()

percent_missing = (total_missing/total_cells) * 100
print(percent_missing)
```

```python
gummybears_2024_sample = gummybears_2024.sample(n = len(gummybears_2024),random_state=1)# take a simple but hier i take all the data as a simple
gummybears_2024_sample
```

```python
# drop null column and rows 
```

```python
gummybears_2024_sample = gummybears_2024_sample.drop(gummybears_2024_sample.columns[0], axis=1) # 1/0 -> row or column
```

```python
gummybears_2024_sample.head()
```

```python
gummybears_2024_sample_final = gummybears_2024_sample.dropna()# drop null value
```

```python
gummybears_2024_sample_final
```

# 3 Data Visualisierung




```python
# Set style for all plots
sns.set(style="whitegrid")

# 1. Balkendiagramm: Anzahl der Gummibärchen pro Farbe
plt.figure(figsize=(10,6))
gummybears_2024_sample_final[['darkred', 'lightred', 'orange', 'yellow', 'green', 'white']].sum().plot(kind='bar', color=['#8B0000', '#FA8072', '#FFA500', '#FFFF00', '#008000', '#FFFFFF'])
plt.title('Anzahl der Gummibärchen pro Farbe')
plt.ylabel('Anzahl')
plt.show()

# 2. Kreisdiagramm: Anteil der Gummibärchenfarben
plt.figure(figsize=(8,8))
gummybears_2024_sample_final[['darkred', 'lightred', 'orange', 'yellow', 'green', 'white']].sum().plot(kind='pie', autopct='%1.1f%%', colors=['#8B0000', '#FA8072', '#FFA500', '#FFFF00', '#008000', '#FFFFFF'], startangle=90)
plt.title('Anteil der Gummibärchenfarben')
plt.ylabel('')
plt.show()

# 3. Boxplot: Verteilung der Körpergröße nach Geschlecht
plt.figure(figsize=(10,6))
sns.boxplot(x='gender', y='height', data=gummybears_2024_sample_final)
plt.title('Verteilung der Körpergröße nach Geschlecht')
plt.xlabel('Geschlecht')
plt.ylabel('Körpergröße (cm)')
plt.show()

# 4. Liniendiagramm: Entwicklung der Gummibärchenanzahl über die Semester
plt.figure(figsize=(10,6))
sns.lineplot(x='semester', y='count_bears', data=gummybears_2024_sample_final, marker='o')
plt.title('Entwicklung der Gummibärchenanzahl über die Semester')
plt.xlabel('Semester')
plt.ylabel('Anzahl Gummibärchen')
plt.show()
```
